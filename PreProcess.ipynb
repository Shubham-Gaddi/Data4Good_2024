## Getting Key Phrases for Q21

```python
import os
from azure.core.credentials import AzureKeyCredential
from azure.ai.textanalytics import TextAnalyticsClient
import time
import pandas as pd

# Initialize the Text Analytics client
text_analytics_client = TextAnalyticsClient(endpoint=endpoint, credential=AzureKeyCredential(key))

# Define rate limits
requests_per_second_limit = 100
requests_per_minute_limit = 300

# Initialize variables
q_21 = []
time_counter = time.time()

# Process the Series in chunks
for i in range(0, len(df_21), 10):
    chunk = df_21[i:i+10].tolist()
    articles = [x for x in chunk if pd.notnull(x)]  # Filter out NaN values

    # Extract key phrases
    result = text_analytics_client.extract_key_phrases(articles)
    for idx, doc in enumerate(result):
        if not doc.is_error:
            print("Key phrases in article #{}: {}".format(
                i + idx + 1,
                ", ".join(doc.key_phrases)
            ))
            q_21.append(doc.key_phrases)

    # Rate limiting logic
    requests_in_this_second = (i // 50) % requests_per_second_limit
    requests_in_this_minute = (i // 50) % requests_per_minute_limit
    
    if requests_in_this_second == 0:
        time_taken_for_batch = time.time() - time_counter
        if time_taken_for_batch < 1:
            time.sleep(1 - time_taken_for_batch)
        time_counter = time.time()
    
    if requests_in_this_minute == 0:
        time_taken_for_minute = time.time() - time_counter
        if time_taken_for_minute < 60:
            time.sleep(60 - time_taken_for_minute)
        time_counter = time.time()
```

```python
total_non_null_q21 = len(q_21)
counter = 0

for i in range(len(df_train)):
    if counter < total_non_null_q21:
        if pd.notnull(df_train.at[i, 'q21']):
            df_train.at[i, 'q21'] = q_21[counter]
            print(q_21[counter])
            counter += 1
        else:
            df_train.at[i, 'q21'] = 'N/A'
            print('N/A')
    else:
        break
```
